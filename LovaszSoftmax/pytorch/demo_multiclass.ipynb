## Batch of 5 random images with classes [0, 1, 2] and void (255)

### Labels

```python
np.random.seed(18)
C = 3
B = 5
H = 200
labels_ = [dummy_triangles(H, [0, 255, 1, 2]) for b in range(B)]
labels = torch.stack(list(map(torch.from_numpy, map(np.array, labels_)))).long()
pil_grid(labels_, 5, margin=1)
```

### Features

```python
np.random.seed(57)
B, H, W = labels.size()
labels_ = labels.clone()
labels_[labels_ == 255] = labels_[labels_ == 255].random_(C) # random feats for void
labels_1hot = torch.zeros(B, C, H, W)
labels_1hot.scatter_(1, labels_.unsqueeze(1), 1);
```

```python
feats = labels_1hot.clone().float()
feats += feats.new(feats.size()).normal_(0, 2) # additive gaussian noise
```

```python
labels = cuda(labels)
feats = Variable(cuda(feats))
```

## Model

### definition

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.conv = nn.Conv2d(C, C, 3, padding=1)
        
    def forward(self, x):
        return x + self.conv(x)
```

```python
m = Model()
init_state_dict = OD([(n, p.clone()) for (n, p) in m.state_dict().items()])
```

```python
m = cuda(m)
```

### Initial prediction

```python
with torch.no_grad():
    _, preds = m(feats).data.max(1)
    print("Initial batch-IoUs:", L.iou(preds.cpu().numpy(), labels.cpu().numpy(), C, ignore=255, per_image=False))
    print("Initial mIoU:", np.mean(L.iou(preds.cpu().numpy(), labels.cpu().numpy(), C, ignore=255, per_image=False)))
```

```python
pil_grid([pil(f.byte().cpu().numpy()) for f in preds], 5, margin=1)
```

## Lov√°sz-Softmax training

```python
m.load_state_dict(init_state_dict)

optimizer = Adam(m.parameters(), lr=0.005)

loss_plot = []
iou_plot = []

for iter in tqdm(range(1000)):
    optimizer.zero_grad()
    out = m(feats)
    out = F.softmax(out)
    loss = L.lovasz_softmax(out, labels, ignore=255)
    loss.backward()
    optimizer.step()
    
    _, preds = out.data.max(1)
    loss_plot.append(loss.item())
    iou_plot.append(L.iou_binary(preds.cpu().numpy(), labels.cpu().numpy(), ignore=255, per_image=False))
```

```python
plt.figure(figsize=(10, 5))
plt.suptitle(u'Lov√°sz-Softmax training')
plt.subplot(1, 2, 1)
plt.plot(loss_plot)
plt.ylabel('loss')
plt.xlabel('iteration')

plt.subplot(1, 2, 2)
plt.plot(iou_plot)
plt.ylabel('Image-IoU (%)')
plt.xlabel('iteration')
```

### Final prediction

```python
with torch.no_grad():
    _, preds = m(feats).data.max(1)
    print("Final batch-IoUs:", L.iou(preds.cpu().numpy(), labels.cpu().numpy(), C, ignore=255, per_image=False))
    print("Final mIoU:", np.mean(L.iou(preds.cpu().numpy(), labels.cpu().numpy(), C, ignore=255, per_image=False)))
```

```python
pil_grid([pil(f.byte().cpu().numpy()) for f in preds], 5, margin=1)
```

## Cross-entropy training

```python
m.load_state_dict(init_state_dict)

optimizer = Adam(m.parameters(), lr=0.005)

loss_plot_x = []
iou_plot_x = []

for iter in tqdm(range(1000)):
    optimizer.zero_grad()
    out = m(feats)
    loss = L.xloss(out, labels, ignore=255)
    loss.backward()
    optimizer.step()
    
    _, preds = out.data.max(1)
    loss_plot_x.append(loss.item())
    iou_plot_x.append(L.iou_binary(preds.cpu().numpy(), labels.cpu().numpy(), ignore=255, per_image=False))
```

```python
plt.figure(figsize=(10, 5))
plt.suptitle(u'Binary cross-entropy training')
plt.subplot(1, 2, 1)
plt.plot(loss_plot_x)
plt.ylabel('loss')
plt.xlabel('iteration')

plt.subplot(1, 2, 2)
plt.plot(iou_plot_x)
plt.ylabel('Image-IoU (%)')
plt.xlabel('iteration')
```

### Final prediction

```python
with torch.no_grad():
    _, preds = m(feats).data.max(1)
    print("Final batch-IoUs:", L.iou(preds.cpu().numpy(), labels.cpu().numpy(), C, ignore=255, per_image=False))
    print("Final mIoU:", np.mean(L.iou(preds.cpu().numpy(), labels.cpu().numpy(), C, ignore=255, per_image=False)))
```
